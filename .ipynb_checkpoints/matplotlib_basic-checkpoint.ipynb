# 06장 matplotlib_basic 

 - 데이터 시각화

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

선그래프

```python
a = np.array([10,15,19,20,25])
plt.plot(a)
b = np.arange(20)
plt.plot(b)
```

```python
# 1차 방정식의 그래프 : f(x) = ax + b

x = np.arange(20)
y = x * 2
[x,y]

plt.plot(x,y)
plt.show()
```

```python
# 2차 방정식의 그래프 : f(x) = ax**2 + b, 포물선의 방정식

x = np.arange(-10,11,1)
y = 2*x** 2
[x,y]
plt.plot(x,y)
plt.plot(x, 2*x)          # 새로운 그래프마다 색깔이 다르게 할당
plt.show()
```

```python
x = np.arange(-5,5,0.1)
y = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y)
plt.plot(x,y2)
plt.show()\
```
