# 학습 (예측)

```python
(1)로그로 변환해서 값을 구하는 함수/공식, 값이 너무 커서, 변화량이 눈에 보임

poly = np.polyfit(years,np.log(counts),deg=1)  # deg = 회귀 다항식의 차수
# polyfit(x,y,deg) : x 값과 y값을 사용하여 기울기를 구함, 변화량이 없으면 기울기가 구하기 힘들어짐
# Least Squeares Polyment fit 
# 선과 데이터의 차이가 가장 적은 선
print('poly',poly)
print(type(poly[0]))                           #  w, 기울기    <class 'numpy.ndarray'>
print(type(poly[1]))                           #  b , y절편
print(poly) 

plt.subplot(2,2,1)
plt.semilogy(years,counts,'o')                 # '0' 산점도로 표현
                                               # 산점도 semilogy가 log값으로 변환해줌

plt.subplot(2,2,2)
plt.semilogy(years,counts)
plt.semilogy(years,np.exp(np.polyval(poly,years)))
# exp : 지수로 바꿈 위에서 log로 바꿨던 값을 다시 지수로 바꾼 후 semilogy로 로그 값으로 다시 만들어줌
plt.show()


# hypothesis 가설/예측 방정식
# f(x)3.61559210e-01 -7.05783195e+02*X + 7654920753.744821


# 예측
(1)의 식을 다시 원래의 값으로 변환하는 식, 이걸해야 원래 원했던 값이 나온다.

predict = np.exp(np.polyval(poly,[2015,2016,2017,2019,2020]))
print('2015년', predict[0])
print('2019년', predict[3])
print('2020년', predict[4])
# polyfit은 직선의 방정식을 구한다.
```
