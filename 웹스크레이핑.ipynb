```python
# 웹크롤링
import urllib.request as REQ
import urllib.parse
```

```python
kurl = 'http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnld=133'
data = REQ.urlopen(kurl)
data=data.read()
x=data.decode('utf-8')              # 한글 변환 
x
```

```python
# url을 설정  : 기상청 
API = 'http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp'
# stnld : 중기예보 지역별코드 3자리
# 
values = {
    'stnId' : '133' # 충청남도    
}

#url 요청시 필요한 param 생성
params = urllib.parse.urlencode(values)
#print(params)  stnld : 133
# param 을 주소 뒤에 ?와 붙여 get 방식으로 요청

url = API + '?' +params
print('url=',url)

# 다운로드
data = REQ.urlopen(url).read()
text = data.decode('utf-8')
print(text)
```

```python
# 웹스크레이핑
from bs4 import BeautifulSoup
response = REQ.urlopen(url)
soup = BeautifulSoup(response,'html.parser')
soup
```

```python
# type(soup)    =    bs4.BeautifulSoup
soup.select('location')
```

```python
for location in  soup.select('location'):
    print(location.city.string)
    print('-'*70)
    for data in location.select('data'):
        print('날짜:',data.tmef.string)
        print('날짜:',data.wf.string)
        print('최저:',data.tmn.string)
        print('최고:',data.tmx.string)  
```
